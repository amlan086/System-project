```python
import cv2
import numpy as np
import os
import math
import numpy as np
from PIL import Image
width = 122
height =96
dim = (width, height)

for frm_no in range(1,101):
    dr='./movie fight/Peliculas/noFights/'+str(frm_no)+'.mpg'
    cap = cv2.VideoCapture(dr)
    fl='ndata_'+str(frm_no)
    try:
        if not os.path.exists(fl):
            os.makedirs(fl)
    except OSError:
        print ('Error: Creating directory of data')

    currentFrame = 0
    while(cap.isOpened()):
        ret, frm = cap.read()
        
        if ret == True:
            frame = cv2.resize(frm, dim, interpolation = cv2.INTER_AREA)
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            #ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
            name = './'+fl+'/frame' + str(currentFrame) + '.jpg'
            #print ('Creating...' + name)
            cv2.imwrite(name, gray)


            currentFrame += 1
        else:
            break

    cap.release()
    cv2.destroyAllWindows()

for frm_no in range(1,101):
    dr='./movie fight/Peliculas/fights/newfi'+str(frm_no)+'.avi'
    cap = cv2.VideoCapture(dr)
    fl='data_'+str(frm_no)
    try:
        if not os.path.exists(fl):
            os.makedirs(fl)
    except OSError:
        print ('Error: Creating directory of data')

    currentFrame = 0
    while(cap.isOpened()):
        ret, frm = cap.read()
        
        if ret == True:
            frame = cv2.resize(frm, dim, interpolation = cv2.INTER_AREA)
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            #ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
            name = './'+fl+'/frame' + str(currentFrame) + '.jpg'
            #print ('Creating...' + name)
            cv2.imwrite(name, gray)


            currentFrame += 1
        else:
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import math
import numpy as np
from PIL import Image
import os
import random
import cv2
X=[]
y=[]
for z in range(1,2):
    print(z)
    if z%6!=0:
        com_list=[]
        
        ss='C:\\Users\\User\\Desktop\\violence_detection\\bin_f\\data_'+str(z)
        os.chdir(ss)
        
        file_list = os.listdir(ss) 
        number_of_files = len(file_list)
        #print(str(z) + " --- "+str(number_of_files))
        #lst=random.sample(range(1,number_of_files), 31)
        #lst.sort()


        for x in range(1,number_of_files+1):
            res=[]
            #print(x)
            s='frame'+str(x)+'.jpg'
            #print(s)
            gray=cv2.imread(s)
            
            cell_size=(8,8)
            block_size=(2,2)
            nbins=9


            winSize = (gray.shape[1] // cell_size[1] * cell_size[1],gray.shape[0] // cell_size[0] * cell_size[0])
            blockSize = (block_size[1] * cell_size[1],block_size[0] * cell_size[0])
            blockStride = (cell_size[1], cell_size[0])
            cellSize = (cell_size[1], cell_size[0])
            nbins = 9
            derivAperture = 1
            winSigma = -1.
            histogramNormType = 0
            L2HysThreshold = 0.2
            gammaCorrection = 1
            nlevels = 64
            signedGradients = True
 
            hog = cv2.HOGDescriptor(winSize,blockSize,blockStride,cellSize,nbins,derivAperture,winSigma,histogramNormType,L2HysThreshold,gammaCorrection,nlevels, signedGradients)

            n_cells = (gray.shape[0] // cell_size[0], gray.shape[1] // cell_size[1])
            hog_feats = hog.compute(gray).ravel()
            hog_lst=hog_feats.tolist()
            X.append(hog_lst)
            y.append(1)
```
