```python
import math
import numpy as np
from PIL import Image
import os
import random
import cv2
X=[]
y=[]
for z in range(1,31):
    print(z)
    if z%6!=0:
        com_list=[]
        
        ss='C:\\Users\\User\\Desktop\\violence_detection\\ff\\data_'+str(z)
        os.chdir(ss)
        
        file_list = os.listdir(ss) 
        number_of_files = len(file_list)
        #print(str(z) + " --- "+str(number_of_files))
        #lst=random.sample(range(1,number_of_files), 31)
        #lst.sort()


        for x in range(number_of_files):
            res=[]
            #print(x)
            s='frame'+str(x)+'.jpg'
            #print(s)
            im = Image.open(s, 'r')
            
            
            hog = cv2.HOGDescriptor()
            im = cv2.imread(s)
            h = hog.compute(im).ravel()
            
            
            
            hog_list= h.tolist()
            i=0
            if x == 0:
                i=0
                for a in hog_list:
                    com_list.insert(i,a)
                    i=i+1   
            else:
                i=0
                for b in hog_list:
                    cm=b
                    res.append(abs(com_list[i]-cm))
                    com_list[i]=cm
                    i=i+1 

                X.append(res)
                y.append(1)
```
